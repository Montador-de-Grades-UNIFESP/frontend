```python
import pandas as pd
```

```python
df = pd.read_csv("GradeUNIFESP2024.csv", sep=";")
display(df)
```

```python
# Adicionei o Termo
class UC:
    def __init__(self, name, turma, professor, hour, day, termo) -> None:
        self.NOME = str(name)
        self.TURMA = str(turma)
        self.PROFESSORES = str(professor)
        self.HORARIO = str(hour)
        self.DIA = str(day)
        self.TERMO = str(termo)

    def __str__(self) -> str:
        return f"Nome: {self.NOME} Turma: {self.TURMA} Horário: {self.HORARIO} Dia: {self.DIA} Termo: {self.TERMO}"
    
```

```python
# Função que cria uma lista de instâncias da classe UC a partir de um DataFrame
def criar_lista_UCs(data_frame):
    ucs: list[UC] = []
    for i, row in data_frame.iterrows():
        ucs.append(UC(row['Disciplina'], row['Turma'], row['Professor'],  row['Horário'], row['Dia da Semana'], row['Termo']))
    return ucs
```

```python
# Função para agrupar as aulas iguais
def agrupar_aulas(ucs):
    # Criando DataFrame a partir da lista de UCs
    df_ucs = pd.DataFrame([vars(uc) for uc in ucs])

    # Agrupando por disciplina, turma e professor e agregando horários e dias
    grouped_ucs = df_ucs.groupby(['NOME', 'TURMA'], as_index=False).agg({
        'HORARIO': lambda x: ','.join(x),
        'DIA': lambda y: ','.join(y),
        'PROFESSORES': 'first',  # Mantendo o professor do primeiro registro
        'TERMO': 'first'  # Mantendo o termo do primeiro registro
    })

    return grouped_ucs
```

```python
lista_UCs = criar_lista_UCs(df)
lista_UCs = agrupar_aulas(lista_UCs)
display(lista_UCs.head(50))
```

```python
# Salvando em um arquivo CSV com delimitador ';'
lista_UCs.to_csv('Materias_2024.csv', index=False, sep=';')
```
